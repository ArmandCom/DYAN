```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def kalman(C, P, y, F, H, R,Q):
    # UPDATE C, P based on measurement y    
    # From Minus status at time t-1 --> Plus status at time t-1
    s = H @ P @ H.T + R  # residual convariance
    K = P @ H.T * np.linalg.inv(s)    # Kalman gain
    C = C + K@(y - H @ C)
    I = np.matrix(np.eye(F.shape[0])) # identity matrix
    P = (I - K@H)@P
    # PREDICT x, P 
    # From Plus status at time t-1 --> Minus status at time t
    C = F@C 
    P = F@P@F.T + Q
    
    return C, P

def gridRing(N):
    epsilon_low = 0.15
    epsilon_high = -0.01
    rmin = (1-epsilon_low)
    rmax = (1+epsilon_high)
    thetaMin = 0
    thetaMax = np.pi/2
    delta = 0.05
    Npole = int(N/4)
    Pool = generateGridPoles(delta,rmin,rmax,thetaMin,thetaMax)
    M = len(Pool)
    idx = random.sample(range(0, M), Npole)
    P = Pool[idx]
    Pall = np.concatenate((P,-P, np.conjugate(P),np.conjugate(-P)),axis = 0)

    return P,Pall
## Generate the grid on poles
def generateGridPoles(delta,rmin,rmax,thetaMin,thetaMax):
    rmin2 = pow(rmin,2)
    rmax2 = pow(rmax,2)
    xv = np.arange(-rmax,rmax,delta)
    x,y = np.meshgrid(xv,xv,sparse = False)
    mask = np.logical_and( np.logical_and(x**2 + y**2 >= rmin2 , x**2 + y **2 <= rmax2),
                            np.logical_and(np.angle(x+1j*y)>=thetaMin, np.angle(x+1j*y)<=thetaMax ))
    px = x[mask]
    py = y[mask]
    P = px + 1j*py
    
    return P
```

```python
## generate synthetic data
P,Pall = gridRing(40)


sysOrder = 2;
Ninit = 10; # Number of trial initial conditions/poles pair
y = np.zeros((Ninit,1000))
p1 = np.array([P[1], np.conjugate(P[1])])
r1 = - np.flipud(np.poly(p1))
    
# Generate random initial conditions for Ninit realizations y
y[:,0:sysOrder] = np.random.uniform(low=0.0, high=1.0, size=(Ninit,sysOrder))
    
# Propagate them to the end of the temporal horizon and save in ydata_clean
for i in range (sysOrder,1000-1):
    y[:,i] = np.dot(y[:,i-sysOrder:i] , r1[0:2])

# Add Noise
nrange = np.absolute(np.amax(y) - np.amin(y))
level = 0.05
print('error level',level * nrange)
ynoise = y + level * nrange * np.random.normal(0, 1, size=y.shape)

#ynoise = y

xbar = np.arange(0,1000)
plt.plot(xbar,y[8,:],xbar,ynoise[8,:])
plt.show()
```

```python
# MINI KF Example
p1 = P[1]
r = abs(p1)
the = np.angle(p1)
T = 4
# generate dictionary
D = np.zeros((T,T))
for i in range (0,T):
    D[i,:] = [np.power(r,i)*np.cos(the*i),np.power(-r,i)*np.cos(the*i),np.power(r,i)*np.sin(the*i),np.power(-r,i)*np.sin(the*i)]

Dinv = np.linalg.inv(D)
yin = ynoise[8,0:T]
yin_true = y[8,0:T]

c = np.linalg.solve(D,yin)#chnage to inverse
chat = np.linalg.solve(D,yin_true)

C_0 = c[:,np.newaxis]
#chat = chat[:,np.newaxis]
#P_0 = np.cov( np.dot((C_0 - chat) , (C_0 - chat).T ))
#P_0 = np.cov((C_0 - chat) @ (C_0 - chat).T )
#P_0 = np.matrix(np.zeros((4,4)))


F = 0.98* np.matrix(np.eye(T))
R = np.matrix(level * nrange)
Q = (level * nrange) * (Dinv @  Dinv.T)
P_0 = Q

y_pre = np.copy(ynoise[8,:])
c_true = np.linalg.solve(D,yin_true)
cold_all = np.matrix(np.zeros((4,1000)))
cnew_all = np.matrix(np.zeros((4,1000)))

for i in range(T,ynoise[8,:].shape[0]):
    H = [np.power(r,i)*np.cos(the*i),np.power(-r,i)*np.cos(the*i),np.power(r,i)*np.sin(the*i),np.power(-r,i)*np.sin(the*i)]
    H = np.matrix(H)
    y_pre[i] = H @ C_0
    #P_cov[i] = ynoise[8,i]-y_pre[i]
    cold_all[:,i] = C_0
    C_0, P_0 = kalman(C_0, P_0, np.matrix(ynoise[8,i]), F, H, R, Q)
    cnew_all[:,i] = C_0
    #print('cerror = ',C_0-cold)
    #print(ynoise[0,i])
```
