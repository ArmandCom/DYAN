```python
def dct2(a):
    return scipy.fftpack.dct( scipy.fftpack.dct( a, axis=0, norm='ortho' ), axis=1, norm='ortho' )

def idct2(a):
    return scipy.fftpack.idct( scipy.fftpack.idct( a, axis=0 , norm='ortho'), axis=1 , norm='ortho')

def process_im(im, desired_sz=(128, 160)):
    target_ds = float(desired_sz[0]) / im.shape[0]
    im = imresize(im, (desired_sz[0], int(np.round(target_ds * im.shape[1]))))
    d = int((im.shape[1] - desired_sz[1]) / 2)
    im = im[:, d:d + desired_sz[1]]
    # im = imutils.resize(im, width=160,height=128)
    return im

def scipwarp(img, u, v):
    M, N, _ = img.shape
    x = np.linspace(0, N - 1, N)
    y = np.linspace(0, M - 1, M)
    x, y = np.meshgrid(x, y)
    x += u
    y += v
    warped = img
    warped[:, :, 0] = scipy.ndimage.map_coordinates(img[:, :, 0], [y.ravel(), x.ravel()], order=1,
                                                    mode='nearest').reshape(img.shape[0], img.shape[1])
    warped[:, :, 1] = scipy.ndimage.map_coordinates(img[:, :, 1], [y.ravel(), x.ravel()], order=1,
                                                    mode='nearest').reshape(img.shape[0], img.shape[1])
    warped[:, :, 2] = scipy.ndimage.map_coordinates(img[:, :, 2], [y.ravel(), x.ravel()], order=1,
                                                    mode='nearest').reshape(img.shape[0], img.shape[1])
    return warped
```

```python
folderList = ['set10V011']
rootDir = '/home/armandcomas/datasets/Kitti_Flows/city-2011_09_26_drive_0084_sync/'

xfra = 128
yfra = 160
szBlock = 8

frames = [each for each in os.listdir(rootDir) if
          each.endswith(('.npy'))]
frames.sort()
dct_all = np.zeros((9, 2*xfra*yfra))
dct_img = np.zeros((9, xfra, yfra, 2))
for n in range(0, 9, 1):
#     if (n%10)==0:
    print(n)
        
    imgname = os.path.join(rootDir, frames[n])
    img = np.load(imgname).astype(float)
    imgsize = img.shape
    dct = np.zeros(imgsize).astype('float32')

    # Do 8x8 DCT on image (in-place)
    for i in r_[:imgsize[0]:szBlock]:
        for j in r_[:imgsize[1]:szBlock]:
            dct[i:(i+szBlock),j:(j+szBlock)] = dct2( img[i:(i+szBlock),j:(j+szBlock)] )
    dct_img[n, :, :, :] = dct
    dct_all[n, :] = np.transpose(dct, (2,0,1)).reshape(-1, 2*xfra*yfra)
plt.pcolormesh(dct_all.T, cmap='RdBu')
plt.colorbar()
# plt.savefig('dct_evolution.png', dpi=200)
plt.show()
```
